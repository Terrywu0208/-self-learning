```python
#用 Python dict 建立 DataFrame
import pandas as pd

dic = {
    "col 1": [1, 2, 3], 
    "col 2": [10, 20, 30],
    "col 3": list('xyz'),
    "col 4": ['a', 'b', 'c'],
    "col 5": pd.Series(range(3))
}
df = pd.DataFrame(dic)
df
```

```python
#改變 DataFrame 裡的欄位名稱
rename_dic = {"col 1": "hi", "col 2": "lala"}
df = df.rename(rename_dic, axis=1)
df
```

```python
#改變 DataFrame 裡的欄位名稱
#list(df.columns[3:]) 將沒改變名字的行補上去
df.columns = ['x(new)', '10x(new)',"asd"] + list(df.columns[3:])
df
```

```python
base_url = "https://data.taipei/api/getDatasetInfo/downloadResource?id=2f238b4f-1b27-4085-93e9-d684ef0e2735&rid=8a319d4b-cadf-436e-8396-d1a356b6c886"
df_raw = pd.read_csv(base_url, encoding='big5')

# 複製一份做處理
df_copy = df_raw.copy()
df_raw
```

```python
df_raw.to_csv("data.csv")
```

```python
df_copy['區序'] = df_copy['區序'].apply(lambda x: ''.join([i for i in x if not i.isdigit()])) #apply 函數  对一个DataFrame进行逐行、逐列和逐元素的操作
df_copy = (df_copy[df_copy['性別'].isin([1, 2])]
      .groupby(['區序', '性別'])[['死亡人數', '受傷人數']]
      .sum()
      .reset_index()
      .sort_values('受傷人數'))

df_copy['性別'] = df_copy['性別'].apply(lambda x: '男性' if x == 1 else '女性')
df_copy = df_copy.reset_index().drop('index', axis=1)

# 顯示結果
df_copy
```
